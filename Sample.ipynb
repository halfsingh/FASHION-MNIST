## Data Distribution

```python
sns.set(color_codes=True)
sns.distplot(train_csv.iloc[:,0],label = 'LABELS',kde=False
             ,color='red',norm_hist=False,rug=False); # PLOTTING DISTRIBUTION OF TRAIN
```

```python
size_of_img = (1,28,28,1)
fig=plt.figure(figsize=(72,72))
for i in range(60):
    ax=fig.add_subplot(12,12,i+1)
    plot_image = np.array(train_csv.iloc[i,1:]).reshape(size_of_img)
    ax.imshow(plot_image,cmap='inferno')
plt.show()
```
